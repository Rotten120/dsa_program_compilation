## Exercise 1: Text Analysis Tool
**Objective:** Practice string manipulation, loops, conditional statements, and functions

### Problem Description
Create a text analysis tool that can analyze a given text and provide various statistics.

### Requirements
1. Create functions to analyze a text string:
   - `count_words(text)`: Count total number of words
   - `count_sentences(text)`: Count sentences (assume sentences end with '.', '!', or '?')
   - `count_paragraphs(text)`: Count paragraphs (separated by double newlines)
   - `most_common_word(text)`: Find the most frequently used word (ignore case)
   - `average_word_length(text)`: Calculate average length of words
   - `find_long_words(text, min_length)`: Find all words longer than specified length

2. Clean the text by removing punctuation when counting words
3. Handle case sensitivity appropriately

### Sample Text

```
Python is a high-level programming language. It is known for its simplicity and readability.
Python supports multiple programming paradigms including procedural, object-oriented, and functional programming.

Many developers choose Python for its extensive libraries and frameworks. 
The language is widely used in web development, data science, artificial intelligence, and automation.
Python's philosophy emphasizes code readability and simplicity!
```

### Expected Output Format

```
=== TEXT ANALYSIS REPORT ===
Total Words: 56
Total Sentences: 6
Total Paragraphs: 2
Most Common Word: "python" (appears 4 times)
Average Word Length: 6.2 characters
Words longer than 8 characters: ['programming', 'language', 'simplicity', 'readability', ...]
```


````python
#your code
class TextAnalysis:
    def __init__(self, text):
        self.text = text #To store the text
        
    def count_words(text):
        word_count = len(text.split())  # To count how many words by using split()
        return word_count
    
    def count_sentences(text):
        sentences = 0
        for char in text:
            if char in ".!?":  # To check if there is any punctuation to see if the sentence ends
                sentences += 1
        return sentences
    
    def count_paragraphs(text):
        paragraphs = text.count("\n\n")  # To find the paragraph breaks
        paragraphs += 1  # adds to the count
        return paragraphs

    def most_common_word(text):
        words = text.split()  
        frequency = {} #To store the word and how much of the word of is repeated

        stopwords = {"and", "to", "the", "is", "a", "of", "in", "on", "for", "with"} #To make sure the connecting words are not counted

        for word in words:
            word = word.lower().strip(".,!?")  # To remove punctuation and make it lowercase

            if word.endswith("'s"):  # To remove possessive 's
                word = word.replace("'s", "")

            if word not in stopwords and word != "":
                if word in frequency:
                    frequency[word] += 1 #Adds the amount in that specific word
                else:
                    frequency[word] = 1

        most_common_word = max(frequency, key=frequency.get) #Gets the most common by using max()
        most_common_word_count = frequency[most_common_word] # To see how many times it appeared
        return most_common_word, most_common_word_count
    
    def average_word_length(text):
        words = TextAnalysis.count_words(text) 
        cleaned_text = text.replace(".", "").replace(",", "").replace("!", "").replace("?", "").replace("-", "")#to remove the punctuations to have a more accurate count of characters
        character_counter = len(cleaned_text.replace(" ", "")) #to get the amount of characters in the text and also used replace to remove the spaces to get a more accurate count
        average_length = character_counter / words #To get the average
        return average_length

    def find_long_words(text):
        long_words_list = []
        words = text.split()  

        for word in words:
            word = word.lower().strip(".,!?")  # To remove punctuation and make it lowercase

            if word.endswith("'s"):  # To remove possessive 's
                word = word.replace("'s", "")

            if len(word) >= 8 and word not in long_words_list: # it checks if the character amount is = or > than 8 and also checks if the word is in the list, so that duplicates are not printed and stored again
                long_words_list.append(word) # stores the word in the list

        return long_words_list
                

def main():
    text = """Python is a high-level programming language. It is known for its simplicity and readability.
Python supports multiple programming paradigms including procedural, object-oriented, and functional programming.

Many developers choose Python for its extensive libraries and frameworks.
The language is widely used in web development, data science, artificial intelligence, and automation.
Python's philosophy emphasizes code readability and simplicity!"""

    word, count = TextAnalysis.most_common_word(text)
    print("=== TOTAL ANALYSIS REPORT ===")
    print("Total Words:", TextAnalysis.count_words(text))
    print("Total Sentences:", TextAnalysis.count_sentences(text))
    print("Total Paragraphs:", TextAnalysis.count_paragraphs(text))
    print(f'Most Common Word: "{word}" (which appears {count} times)')
    print("Average Word Length:", TextAnalysis.average_word_length(text))
    print("Words longer than 8 characters:", TextAnalysis.find_long_words(text))

if __name__ == "__main__":
    main()
````

## Exercise 2: Student Grade Calculator
**Objective:** Practice using variables, lists, functions, and conditional statements

### Problem Description
Create a program that calculates the final grade for students in a class. The program should:

1. Store student names and their scores in appropriate data structures
2. Calculate the average score for each student
3. Determine the letter grade based on the average
4. Display a formatted report

### Requirements
- Use a list to store student names: `["Alice", "Bob", "Charlie", "Diana", "Eve"]`
- Each student has 4 test scores. Store these in a nested list structure
- Create a function `calculate_average(scores)` that returns the average of a list of scores
- Create a function `get_letter_grade(average)` that returns:
  - 'A' for 90-100
  - 'B' for 80-89
  - 'C' for 70-79
  - 'D' for 60-69
  - 'F' for below 60
- Display results in a formatted table

### Sample Data
```
students = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
scores = [
    [85, 92, 78, 94],    # Alice's scores
    [76, 83, 91, 87],    # Bob's scores
    [94, 89, 96, 93],    # Charlie's scores
    [67, 74, 82, 79],    # Diana's scores
    [88, 85, 90, 92]     # Eve's scores
]
```

### Expected Output
```
=== STUDENT GRADE REPORT ===
Alice:    Average: 87.25, Grade: B
Bob:      Average: 84.25, Grade: B
Charlie:  Average: 93.00, Grade: A
Diana:    Average: 75.50, Grade: C
Eve:      Average: 88.75, Grade: B
```


````python
#your code
class GradeCalculator:
    def __init__(self, scores, students):
        self.scores = scores #To store the scores 
        self.students = students #To store the students from the list

    def calculate_average(scores): 
        return sum(scores) / len(scores) #It adds them up using sum() and dividing it by getting how many grades are there by using len()
    
    def get_letter_grade(average):
            if 90 <= average <= 100: #Using 2 numbers as a range to get more of a accurate result
                letter_grade = "A"
                return letter_grade
        
            elif 80 <= average <= 89:
                letter_grade = "B"
                return letter_grade

            elif 70 <= average <= 79:
                letter_grade = "C"
                return letter_grade

            elif 60 <= average <= 69:
                letter_grade = "D"
                return letter_grade
            
            else:
                letter_grade = "F"
                return letter_grade

def __main__():
    students = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
    scores = [
        [85, 92, 78, 94],  # Alice's scores
        [76, 83, 91, 87],  # Bob's scores
        [94, 89, 96, 93],  # Charlie's scores
        [67, 74, 82, 79],  # Diana's scores
        [88, 85, 90, 92]   # Eve's scores
    ]
    print("=== STUDENT GRADE REPORT ===")
    for i in range(len(students)): #To have it link up the student with their respective grade and letter grade 
        average = GradeCalculator.calculate_average(scores[i])  # Gets the avearage
        grade = GradeCalculator.get_letter_grade(average)       # and then gets the letter grade
        print(f"{students[i]}:{' ' * (7 - len(students[i]))} Average: {average:<6.2f}, Grade: {grade}") #Used {' ' * (7 - len(students[i]))} to correctly have the right colon placement,amount of space, and aligns, which :<6 also aligns it but now between average and grade

if __name__ == "__main__":
    __main__()
````
